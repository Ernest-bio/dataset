## 5. Анализ числовых признаков

Оценим распределения, выбросы и основные статистики для `length_m`, `weight_kg`, `height_m`. Будем использовать IQR и z-score для оценки выбросов.

```python
num_cols = ['length_m','weight_kg','height_m']
existing_num = [c for c in num_cols if c in df.columns]
df[existing_num].describe().T
```

```python
from scipy import stats

def iqr_outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return series[(series < lower) | (series > upper)]

def zscore_outliers(series, thresh=3):
    zs = np.abs(stats.zscore(series.dropna()))
    return series.dropna()[zs > thresh]

for c in existing_num:
    s = df[c].dropna()
    print(f'-- {c} --')
    print('count:', s.count())
    print('mean, median, std:', s.mean(), s.median(), s.std())
    iqr_o = iqr_outliers(s)
    z_o = zscore_outliers(s)
    print('IQR выбросов:', len(iqr_o), ' (', round(len(iqr_o)/len(s)*100,2), '% )')
    print('Z-score выбросов:', len(z_o), ' (', round(len(z_o)/len(s)*100,2), '% )\n')
```

## 8. Выявление аномалий и необычных строк

Попробуем собрать строки с любым из признаков, где есть серьёзные выбросы или логические несоответствия (напр., вес=0 или отрицательные значения).

```python
anomaly_mask = pd.Series(False, index=df.index)
for c in existing_num:
    s = df[c]
    anomaly_mask = anomaly_mask | (s <= 0) | (s.isna() & False)  

for c in existing_num:
    s = df[c]
    if s.dropna().shape[0] > 0:
        zs = np.abs(stats.zscore(s.dropna()))
        out_idx = s.dropna().index[zs > 4]  
        anomaly_mask.loc[out_idx] = True

print('Найдено подозрительных строк (по числовым аномалиям):', anomaly_mask.sum())
display(df[anomaly_mask].head(20))
```

## 9. Сводные метрики по оценкам

Здесь собраны числовые метрики по каждой оценке: полнота, уникальность, выбросы, дубликаты.

```python
metrics = {}
metrics['n_rows'] = len(df)
metrics['n_cols'] = df.shape[1]
metrics['n_duplicates'] = int(df.duplicated().sum())
metrics['pct_duplicates'] = df.duplicated().mean() * 100
metrics['missing_perc_overall'] = df.isna().mean().mean() * 100
metrics['columns_with_missing'] = int((df.isna().sum() > 0).sum())

for c in existing_num:
    s = df[c].dropna()
    if len(s) == 0:
        metrics[f'{c}_iqr_outlier_pct'] = np.nan
    else:
        metrics[f'{c}_iqr_outlier_pct'] = len(iqr_outliers(s)) / len(s) * 100

pd.Series(metrics).to_frame('value')
```
